```python
def get_dateRange(beginDate, endDate):
    '''
    :param beginDate: 时间范围开始的日期
    :param endDate: 时间范围结束的日期
    :return: 该时间范围内的时间序列
    '''
    dates = []
    dt = datetime.datetime.strptime(beginDate, "%Y-%m-%d")
    date = beginDate[:]
    while date <= endDate:
        dates.append(date)
        dt = dt + datetime.timedelta(1)
        date = dt.strftime("%Y-%m-%d")
    return dates
```

```python
def get_movie_data_in_daterange(data, date_range):
    '''
        获取一定时间范围内，某部电影的数据
    '''
    weekday = []
    for i in date_range:
        weekday.append(data[data['date'] == i])
    return weekday
```

#### 4.1 “time_interval”，“时间”与“影片上上映时间”的差值

```python
def get_time_interval(x, y):
    '''
        获取影片已经上映多久；x:现在时间； y:影片上映时间；
    '''
    from dateutil.parser import parse
    a = parse(x)
    b = parse(y)
    return (a - b).total_seconds()
```

```python
new_test_data['time_interval'] = new_test_data.apply(
        lambda row: get_time_interval(row['时间'], row['影片上线时间']), axis=1)
```

```python
def transfer0to1(data, name_type, name_list, dic):
    '''
        将新增特征列中，将属于高频导演/主演的特征，从0转成1
        "name_type":“导演”或“主演”
        "name_list": 高频导演/主演的总名单
        "dic": 新特征列名与数字索引对应的字典
    '''
    for num, item in enumerate(data[name_type]):

        if item and len(str(item).split(",")) == 1:
            if item in name_list:
                data.iloc[num, dic[item]] = 1

        if item and len(str(item).split(",")) > 1:
            for n in item.split(","):
                if n in name_list:
                    data.iloc[num, dic[n]] = 1
                else:
                    continue
```

#### 4.3 is_weekday”判断是否是工作日

```python
def get_weekday(x):
    import calendar
    '''
        将日期转化为对应的星期数
    '''
    from datetime import datetime
    dateString = x.split()[0]
    week_day = datetime.strptime(dateString, "%Y-%m-%d").weekday()
    return calendar.day_name[week_day]


def is_weekday(day):
    '''
    判断是否是工作日,是为1，不是为0。
    '''
    if day == "Saturday" or day == "Sunday":
        return 0
    else:
        return 1
```

```python
new_test_data['weekday'] = new_test_data['时间'].apply(get_weekday)
new_test_data['is_weekday'] = new_test_data['weekday'].apply(is_weekday)
```

#### 4.7 时间区间特征

```python
def get_hour(x):
    '''
        用“时间”的时间戳补充新特征小时"hour"
    '''
    return x.split()[1].split(":")[0]

def get_time_partition(x):
    '''
        将时间点转化为对应的时间区间
    '''
    x = int(x)
    if x in range(1, 11):   # 1:00-10:00
        return "0"
    if x in range(11, 21):  # 11:00-20:00
        return "1"
    else:                   # 21:00-0:00
        return "2"
```

```python
def transfer0to1(data, name_type, name_list, dic):
    '''
        将新增特征列中，将属于高频导演/主演的特征，从0转成1
        "name_type":“导演”或“主演”
        "name_list": 高频导演/主演的总名单
        "dic": 新特征列名与数字索引对应的字典
    '''
    for num, item in enumerate(data[name_type]):

        if item and len(str(item).split(",")) == 1:
            if item in name_list:
                data.iloc[num, dic[item]] = 1

        if item and len(str(item).split(",")) > 1:
            for n in item.split(","):
                if n in name_list:
                    data.iloc[num, dic[n]] = 1
                else:
                    continue
```

#### 4.9 以6为时间窗口，获取每6条“当前热度值”的平均值

```python
# new_test_data.index = [i for i in range(0, 19276 )]
```

```python
def MaxMinNormalization(x):
    '''
        "当前热度值"归一化处理，0为该列min,6304为该列max
    '''

    x = (float(x) - 0) / (6304 - 0)
    return x
```
